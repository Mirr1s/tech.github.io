```python

def VMtranslator (archivo):

  nombreA = archivo.strip(".vm")
  nuevo = open(nombreA+".asm","x")

  k=0

  m1file = open(archivo)
    #Se eliminan espacios en blanco y saltos de l√≠nea
  for Olinea in m1file.readlines():
    Olinea = Olinea.strip(" ")
    Olinea = Olinea.rstrip("\n")


    if Olinea.startswith("push"):

      numero1 = Olinea.rsplit(" ")
      num1=int(numero1[2])

      res1=16+num1;
      static1 = str(res1)

      restemp=5+num1;
      temp1 = str(restemp)

      if "constant" in Olinea:
        nuevo.write("\n@" + numero1[2])
        nuevo.write("\nD=A")
        nuevo.write("\n@SP")
        nuevo.write("\nA=M")
        nuevo.write("\nM=D")
        nuevo.write("\n@SP")
        nuevo.write("\nM=M+1")
      if "pointer" in Olinea:
        if numero1[2] == "1":
          nuevo.write("\n@THAT")
          nuevo.write("\nD=M")
          nuevo.write("\n@SP")
          nuevo.write("\nA=M")
          nuevo.write("\nM=D")
          nuevo.write("\n@SP")
          nuevo.write("\nM=M+1")
        if numero1[2] == "0":
          nuevo.write("\n@THIS")
          nuevo.write("\nD=M")
          nuevo.write("\n@SP")
          nuevo.write("\nA=M")
          nuevo.write("\nM=D")
          nuevo.write("\n@SP")
          nuevo.write("\nM=M+1")
      if not "constant" in Olinea and not "pointer" in Olinea:
        if "local" in Olinea:
          nuevo.write("\n@LCL")
        if "argument" in Olinea:
          nuevo.write("\n@ARG")
        if "this" in Olinea:
          nuevo.write("\n@THIS")
        if "that" in Olinea:
          nuevo.write("\n@THAT")
        if "static" in Olinea:
          nuevo.write("\n@"+ numero1[2])
        if "temp" in Olinea:
          nuevo.write("\n@"+ numero1[2])
        nuevo.write("\nD=M")
        if "static" in Olinea:
          nuevo.write("\n@"+static1)
        if "temp" in Olinea:
          nuevo.write("\n@"+temp1)
        if not "temp" in Olinea and not "static" in Olinea:
          nuevo.write("\n@"+ numero1[2])
        nuevo.write("\nA=D+A")
        nuevo.write("\nD=M")
        nuevo.write("\n@SP")
        nuevo.write("\nA=M")
        nuevo.write("\nM=D")
        nuevo.write("\n@SP")
        nuevo.write("\nM=M+1")

    if Olinea.startswith("pop"):
      numero1 = Olinea.rsplit(" ")
      num1=int(numero1[2])

      res1=16+num1;
      static1 = str(res1)

      restemp=5+num1;
      temp1 = str(restemp)

      if "local" in Olinea:
        nuevo.write("\n@LCL")
      if "argument" in Olinea:
        nuevo.write("\n@ARG")
      if "this" in Olinea:
        nuevo.write("\n@THIS")
      if "that" in Olinea:
        nuevo.write("\n@THAT")
      if "static" in Olinea:
        nuevo.write("\n@"+ numero1[2])
      if "temp" in Olinea:
        nuevo.write("\n@"+ numero1[2])
      if "pointer" in Olinea:
          if numero1[2] == "1":
              nuevo.write("\n@THAT")
          if numero1[2] == "0":
              nuevo.write("\n@THIS")
      nuevo.write("\nD=M")
      if "static" in Olinea:
        nuevo.write("\n@"+static1)
      if "temp" in Olinea:
        nuevo.write("\n@"+temp1)
      if "pointer" in Olinea:
          if numero1[2] == "1":
              nuevo.write("\n@4")
          if numero1[2] == "0":
              nuevo.write("\n@3")
      if not "pointer" in Olinea and not "temp" in Olinea and not "static" in Olinea:
        nuevo.write("\n@"+ numero1[2])
      nuevo.write("\nD=D+A")
      nuevo.write("\n@R13")
      nuevo.write("\nM=D")
      nuevo.write("\n@SP")
      nuevo.write("\nAM=M-1")
      nuevo.write("\nD=M")
      nuevo.write("\n@R13")
      nuevo.write("\nA=M")
      nuevo.write("\nM=D")

    if Olinea.startswith("add") | Olinea.startswith("sub"):

      nuevo.write("\n@SP")
      nuevo.write("\nAM=M-1")
      nuevo.write("\nD=M")
      nuevo.write("\nA=A-1")
      if "add" in Olinea:
        nuevo.write("\nM=D+M")
      if "sub" in Olinea:
        nuevo.write("\nM=M-D")

    if Olinea.startswith("neg"):
      nuevo.write("\n@SP")
      nuevo.write("\nA=M")
      nuevo.write("\nA=A-1")
      nuevo.write("\nM=-M")


    if Olinea.startswith("eq") | Olinea.startswith("gt") | Olinea.startswith("lt"):
      k=k+1;
      i=str(k)

      nuevo.write("\n@SP")
      nuevo.write("\nAM=M-1")
      nuevo.write("\nD=M")
      nuevo.write("\n@SP")
      nuevo.write("\nAM=M-1")
      nuevo.write("\nD=M-D")
      nuevo.write("\n@True"+i)
      if "eq" in Olinea:
        nuevo.write("\nD;JEQ")
      if "gt" in Olinea:
        nuevo.write("\nD;JGT")
      if "lt" in Olinea:
        nuevo.write("\nD;JLT")
      nuevo.write("\nD=0")
      nuevo.write("\n@False"+i)
      nuevo.write("\n0;JMP")
      nuevo.write("\n(True"+i+")")
      nuevo.write("\nD=-1")
      nuevo.write("\n(False"+i+")")
      nuevo.write("\n@SP")
      nuevo.write("\nA=M")
      nuevo.write("\nM=D")
      nuevo.write("\n@SP")
      nuevo.write("\nM=M+1")

    if Olinea.startswith("and") | Olinea.startswith("or"):
      nuevo.write("\n@SP")
      nuevo.write("\nAM=M-1")
      nuevo.write("\nD=M")
      nuevo.write("\nA=A-1")
      if "and" in Olinea:
        nuevo.write("\nM=D&M")
      if "or" in Olinea:
        nuevo.write("\nM=D|M")

    if Olinea.startswith("not"):
      nuevo.write("\n@SP")
      nuevo.write("\nA=M")
      nuevo.write("\nA=A-1")
      nuevo.write("\nM=!M")


VMtranslator("SimpleAdd.vm")
VMtranslator("StackTest.vm")
VMtranslator("BasicTest.vm")
VMtranslator("PointerTest.vm")
VMtranslator("StaticTest.vm")
```

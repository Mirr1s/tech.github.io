```python
def ensamblador (archivo):
  #Diccionarios comp, dest, jump y simbPredef (Contiene los símbolos predefinidos)

  Comp={
    "0":"0101010",
    "1":"0111111",
    "-1":"0111010",
    "D":"0001100",
    "A":"0110000","M":"1110000",
    "!D":"0001101",
    "!A":"0110001", "!M":"1110001",
    "-D":"0001111",
    "-A":"0110011", "-M":"1110011",
    "D+1":"0011111",
    "A+1":"0110111","M+1":"1011111",
    "D-1":"0001110",
    "A-1":"0110010","M-1":"1110010",
    "D+A":"0000010","D+M":"1000010",
    "D-A":"0010011","D-M":"1010011",
    "A-D":"0000111","M-D":"1000111",
    "D&A":"0000000","D&M":"1000000",
    "D|A":"0010101","D|M":"1010101"
    }
  Dest = {"M":"001","D":"010","MD":"011","A":"100","AM":"101","AD":"110","AMD":"111"}
  Jump = {"JGT":"001","JEQ":"010","JGE":"011","JLT":"100","JNE":"101","JLE":"110","JMP":"111"}
  simbPredef={
    "R0":0,"R1":1,"R2":2,"R3":3,"R4":4,"R5":5,"R6":6,"R7":7,"R8":8,"R9":9,"R10":10,
    "R11":11 ,"R12":12  ,"R13":13  ,"R14":14  ,"R15":15  ,"R16":16,
    "SCREEN":16384,"KBD":24576,"SP":0,"LCL":1,"ARG":2,"THIS":3,"THAT":4,
    "STOP":18,"END":22}
  i = 0;
  j = 0;
  k = 0;
  vD = 0;
  var = 16
  fl = 0;
  m1file = open(archivo)
    #Se eliminan espacios en blanco y saltos de línea
  for Olinea in m1file.readlines():
    Olinea = Olinea.strip(" ")
    Olinea = Olinea.rstrip("\n")

    if Olinea.startswith("/"):
      fl = fl+1;

    if Olinea == "":
      fl = fl+1
    #Se asignan en memoria las constantes, las cuales se reconocen por estar entre paréntesis
    if j >= 0:
      if Olinea.startswith("("):
        st = Olinea.lstrip("(")
        st = st.rstrip(")")
        if st in simbPredef:
          j=j
        else:
          simbPredef[st] = vD-fl
      else:
        vD = vD+1
    print(Olinea)
  print("------------------------------")
  m2file = open(archivo)
  for lin in m2file.readlines():
      lin = lin.strip(" ")
      lin = lin.rstrip("\n")
      #Se encuentran las intrucciones de direccionamiento A, las cuales empiezan con @
      if k>=1:
        Ainstr2 = lin.startswith("@")
        if Ainstr2 == True:
          varlop = 0
          fnt = lin[1:] in simbPredef
          #Se verifica si están predefinidas
          if fnt:
            k=k
          else:
            try:
              a = int(lin[1:])*2
            except:
              simbPredef[lin[1:]] = var
              var = var+1
      k=k+1
  print("------------------------------")
  mfile = open(archivo)
  for linea in mfile.readlines():
    linea = linea.strip(" ")
    linea = linea.rstrip("\n")
    linea = linea.lstrip(" '")
    if i >= 7:
      Ainstr = linea.startswith("@")
      if Ainstr == True:
        try:
          print("0"+bin(int(linea[1:]))[2:].zfill(15))
        except:
          print("0"+bin(int(simbPredef[linea[1:].strip()]))[2:].zfill(15))
      elif linea.startswith("("):
        k=k
      #Se eliminan etiquetas
      else:
          linea = linea.split("//")
          word = linea[0].split("=")
          word2 = linea[0].split(";")

          try: # dest = comp ; jump
            com1 = word[1].split(";")
            dest = Dest[word[0].rstrip()]
            com = com1[0]
          except: # comp ; jmp
            dest = "000"
            com1 = word[0].split(";")
            com = com1[0]
          try:
              jmp = Jump[word2[1].rstrip()]
          except:
              jmp = "000"
          comf = com.rstrip()
          try:
            print("111"+ Comp[comf]  + dest + jmp)
          except:
            k=k
    i=i+1;
  mfile.close()

ensamblador("/Add.asm")
#ensamblador("/Max.asm")
#ensamblador("/MaxL.asm")
#ensamblador("/Pong.asm")
#ensamblador("/PongL.asm")
#ensamblador("/Rect.asm")
#ensamblador("/RectL.asm")
```
